```python
import pandas as pd

# load the csv file (Monthly average air temperatures of the city of Barcelona since 1780) into a pandas data frame
df_history_score = pd.read_csv('/content/sample_data/User_id_2cb32af6-acd1-43e1-91fe-db8e3b695ff5_scorehistory.csv', encoding='utf-8')

# visualize first 5 rows
df_history_score.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(15,15))
plt.plot(df_history_score['on_date'],df_history_score['healthscore'])
plt.xticks(rotation=45)
plt.xlabel("Date")
plt.ylabel("HealthScore")
plt.title("Histoical Health Score Chart")
```

```python
df_history_score['EMA-0.05-HealthScore'] = df_history_score.healthscore.ewm(alpha=0.05, adjust=False).mean()
# mean() , var() variance, std() standard deviation, corr() correlation, cov() covariance
# Higher the Alpha, the more sensitive it is to recent health scores.
# Lower the Alpha, the more it will rely on historical scores. So the health score will change slowly
# higher the health score will change drastically
```
